```python
import time
import random
import matplotlib.pyplot as plt

# Linear Search
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

# Binary Search
def binary_search(arr, x):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid+1
        else:
            high = mid-1
    return -1

sizes = [100, 500, 1000, 5000, 10000]

linear_times = []
binary_times = []

for n in sizes:
    arr = list(range(n))  # sorted for binary search
    
    # Worst case (element not present)
    target = -1  
    
    # Linear Search time
    start = time.time()
    linear_search(arr, target)
    end = time.time()
    linear_times.append(end - start)
    
    # Binary Search time
    start = time.time()
    binary_search(arr, target)
    end = time.time()
    binary_times.append(end - start)

# Plot
plt.plot(sizes, linear_times, label="Linear Search")
plt.plot(sizes, binary_times, label="Binary Search")

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Linear vs Binary Search (Worst Case)")
plt.legend()
plt.grid()

plt.show()
```

# task-3

```python
import time
import matplotlib.pyplot as plt
```

```python
fact_calls = 0

def factorial(n):
    global fact_calls
    fact_calls += 1
    
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
```

```python
fib_calls = 0

def fib(n):
    global fib_calls
    fib_calls += 1
    
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
```

```python
fib_dp_calls = 0

def fib_dp(n):
    global fib_dp_calls
    fib_dp_calls += 1
    
    dp = [0, 1]
    for i in range(2, n+1):
        fib_dp_calls += 1
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]
```

```python
sizes = [5, 10, 15, 20]
```

```python
fact_times = []
fib_times = []
fib_dp_times = []

fact_call_list = []
fib_call_list = []
fib_dp_call_list = []

for n in sizes:
    
    # Factorial
    fact_calls = 0
    start = time.time()
    factorial(n)
    fact_times.append(time.time() - start)
    fact_call_list.append(fact_calls)
    
    # Fibonacci (naive)
    fib_calls = 0
    start = time.time()
    fib(n)
    fib_times.append(time.time() - start)
    fib_call_list.append(fib_calls)
    
    # Fibonacci (DP)
    fib_dp_calls = 0
    start = time.time()
    fib_dp(n)
    fib_dp_times.append(time.time() - start)
    fib_dp_call_list.append(fib_dp_calls)
```

```python
print("n | Fact Calls | Fib Calls | Fib_DP Calls")
for i in range(len(sizes)):
    print(f"{sizes[i]} | {fact_call_list[i]} | {fib_call_list[i]} | {fib_dp_call_list[i]}")
```

```python
plt.plot(sizes, fact_times, label="Factorial")
plt.plot(sizes, fib_times, label="Fibonacci (Naive)")
plt.plot(sizes, fib_dp_times, label="Fibonacci (DP)")

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Execution Time Comparison")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.plot(sizes, fact_call_list, label="Factorial Calls")
plt.plot(sizes, fib_call_list, label="Fib Calls")
plt.plot(sizes, fib_dp_call_list, label="Fib DP Calls")

plt.xlabel("Input Size (n)")
plt.ylabel("Number of Calls")
plt.title("Function Calls Comparison")
plt.legend()
plt.grid()

plt.show()
```

# task-4

```python
t1_calls = 0

def T1(n):
    global t1_calls
    t1_calls += 1
    
    if n <= 1:
        return 1
    
    return T1(n//2) + n
```

```python
t2_calls = 0

def T2(n):
    global t2_calls
    t2_calls += 1
    
    if n <= 1:
        return 1
    
    return T2(n//2) + T2(n//2) + n
```

```python
sizes = [2, 4, 8, 16, 32, 64, 128]
```

```python
t1_times = []
t2_times = []

t1_call_list = []
t2_call_list = []

for n in sizes:
    
    # T1
    t1_calls = 0
    start = time.time()
    T1(n)
    t1_times.append(time.time() - start)
    t1_call_list.append(t1_calls)
    
    # T2
    t2_calls = 0
    start = time.time()
    T2(n)
    t2_times.append(time.time() - start)
    t2_call_list.append(t2_calls)
```

```python
print("n | T1 Calls | T2 Calls")
for i in range(len(sizes)):
    print(f"{sizes[i]} | {t1_call_list[i]} | {t2_call_list[i]}")
```

```python
plt.plot(sizes, t1_times, marker='o', label="T(n)=T(n/2)+n")
plt.plot(sizes, t2_times, marker='o', label="T(n)=2T(n/2)+n")

plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Recurrence Time Comparison")
plt.legend()
plt.grid()

plt.show()
```
